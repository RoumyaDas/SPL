```python
venues = []
col = 'venue'
dfs = [inn_1, inn_2]

for df in dfs:
    ls = df[col].unique().tolist()
    venues.append(ls)
    
venues_list = list(set(element for sublist in venues for element in sublist))

len(venues_list)
```
